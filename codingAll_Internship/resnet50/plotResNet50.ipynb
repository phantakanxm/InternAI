```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the training log
# Using the uploaded file name "training_log.csv"
df_training_log = pd.read_csv("/Users/phantakan/convolution/internai/log/training_log.csv")

# --- Plot 1: Training Loss vs. Epoch ---
plt.figure(figsize=(10, 6))
plt.plot(df_training_log['epoch'], df_training_log['loss'], label='Training Loss', marker='o')
plt.title('Training Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.savefig("training_loss_vs_epoch_en.png")
plt.show()

# --- Plot 2: Validation Loss vs. Epoch ---
plt.figure(figsize=(10, 6))
plt.plot(df_training_log['epoch'], df_training_log['val_loss'], label='Validation Loss', marker='o', color='orange')
plt.title('Validation Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.legend()
plt.grid(True)
plt.savefig("validation_loss_vs_epoch_en.png")
plt.show()

# --- Plot 3: Training Accuracy vs. Epoch ---
plt.figure(figsize=(10, 6))
plt.plot(df_training_log['epoch'], df_training_log['accuracy'], label='Training Accuracy', marker='o', color='green')
plt.title('Training Accuracy vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.savefig("training_accuracy_vs_epoch_en.png")
plt.show()

# --- Plot 4: Validation Accuracy vs. Epoch ---
plt.figure(figsize=(10, 6))
plt.plot(df_training_log['epoch'], df_training_log['val_accuracy'], label='Validation Accuracy', marker='o', color='red')
plt.title('Validation Accuracy vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.grid(True)
plt.savefig("validation_accuracy_vs_epoch_en.png")
plt.show()

# --- Plot 5: Combined Training and Validation Loss vs. Epoch ---
plt.figure(figsize=(12, 7))
plt.plot(df_training_log['epoch'], df_training_log['loss'], label='Training Loss', marker='o')
plt.plot(df_training_log['epoch'], df_training_log['val_loss'], label='Validation Loss', marker='x', linestyle='--')
plt.title('Training and Validation Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.savefig("combined_loss_vs_epoch_en.png")
plt.show()

# --- Plot 6: Combined Training and Validation Accuracy vs. Epoch ---
plt.figure(figsize=(12, 7))
plt.plot(df_training_log['epoch'], df_training_log['accuracy'], label='Training Accuracy', marker='o')
plt.plot(df_training_log['epoch'], df_training_log['val_accuracy'], label='Validation Accuracy', marker='x', linestyle='--')
plt.title('Training and Validation Accuracy vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.savefig("combined_accuracy_vs_epoch_en.png")
plt.show()

print("Graphs generated successfully and saved as PNG files with English labels.")
```
